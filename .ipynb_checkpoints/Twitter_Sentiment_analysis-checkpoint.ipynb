Next step is to remove the punctutions and stopwords from the tweets to further clean the data

```python
import string
string.punctuation
```

```python
import nltk

nltk.download('stopwords')
from nltk.corpus import stopwords


stopwords.words("english")
```

We can define a pipeline to remove the punctuations and the stopwords from each tweet by using a function

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def msg_cleaning(message):
    Test_punc_removed = [char for  char in message if char not in string.punctuation]
    Test_punc__removed_joined =  "".join(Test_punc_removed)
    Test_punc_removed_joined_cleaned = [word for word in Test_punc__removed_joined.split() if word.lower() not in stopwords.words('english')]
    return Test_punc_removed_joined_cleaned
```

We then vectorize the words into an array and use that to train and test our data

```python
vectorizer =  CountVectorizer(analyzer=msg_cleaning,dtype=np.uint8)
tweets_count_vectorizer = vectorizer.fit_transform(data["tweet"])
```

```python
print(tweets_count_vectorizer.toarray())
```

```python
tweets_count_vectorizer.shape
```

```python
X = pd.DataFrame(tweets_count_vectorizer.toarray())
```

```python
X
```

```python
y = data["label"]
```

```python
X.shape
```

```python
y.shape
```

We can utilize the inbuilt naive bayes model from  scikit learn to train on the data available and test it as well

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
NBClassifier = MultinomialNB()
NBClassifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_predict = NBClassifier.predict(X_test)
cm = confusion_matrix(y_test,y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_predict))
```

We can also use the random forest classifier from sklearn to train our model and test it out on the data

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier()
```

```python
rf_classifier.fit(X_train,y_train)
```

```python
y_pred_rf = rf_classifier.predict(X_test)
cm = confusion_matrix(y_test,y_pred_rf)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_pred_rf))
```
